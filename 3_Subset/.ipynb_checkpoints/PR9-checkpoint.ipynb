```python
# e) Melting Data to long format

melted_df = pd.melt(df, id_vars=['Type', 'Category'], value_vars=['comment', 'like', 'share'])
print("\nMelted Data (Long Format)\n", melted_df.head())
```

```python
# f) Casting (pivoting) data to wide format

pivot_df = melted_df.pivot_table(index=['Type', 'Category'], columns='variable', values='value', aggfunc='sum')
print("\nPivoted Data (Wide Format)\n", pivot_df.head())
```
