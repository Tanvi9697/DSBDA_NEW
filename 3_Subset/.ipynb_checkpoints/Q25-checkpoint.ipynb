```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("hepatitis.csv")
```

```python
df.head()
```

```python
df.columns = ['class', 'age', 'sex', 'steroid', 'antivirals', 'fatigue', 'malaise', 'anorexia',
           'liver_big', 'liver_firm', 'spleen_palpable', 'spiders', 'ascites', 'varices',
           'bilirubin', 'alk_phosphate', 'sgot', 'albumin', 'protime', 'histology']
df.head()
```

```python
df.replace('?', np.nan,inplace=True)
```

```python
numeric_cols = ['age', 'bilirubin', 'alk_phosphate', 'sgot', 'albumin', 'protime']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
male_subset = df[df['sex'] == 1]
male_subset.head()
```

```python
female_subset = df[df['sex'] == 2]
female_subset.head()
```

```python
df.shape
```

```python
male_subset.shape
```

```python
female_subset.shape
```

```python
merged_subsets = pd.concat([male_subset, female_subset])
merged_subsets
```

```python
sorted_df = df.sort_values(by=['age', 'sgot', 'protime'])
sorted_df
```

```python
transposed_df = df.transpose()
transposed_df
```

```python
melted_df = pd.melt(df, id_vars=['sex'],
                    value_vars=['age', 'sgot', 'protime', 'bilirubin', 'albumin'],
                    var_name='Metric', value_name='value')
melted_df.head()
```

```python
wide_df = melted_df.pivot_table(index='sex', columns='Metric', values='value', aggfunc='mean')
wide_df = wide_df.reset_index()
wide_df.head()
```
