```python
!pip install lxml
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.meesho.com/women-men-kids-footwear/clp/15mz"
r=requests.get(url)
print(r)
```

```python
soup=BeautifulSoup(r.text,"lxml")
print(soup)
```

```python
Product_Name=[]
names=soup.find_all("p",class_="sc-eDvSVe gQDOBc NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ejhQZU NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ejhQZU")
for i in names:
    name=i.text
    Product_Name.append(name)
    print(Product_Name)
```

```python
Price=[]
prices=soup.find_all("h5",class_="sc-eDvSVe dwCrSh")
for i in prices:
    price=i.text
    Price.append(price)
    print(Price)
```

```python
df=pd.DataFrame({"product name":Product_Name,"price":Price})
```
