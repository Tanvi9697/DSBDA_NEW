```python
url="https://www.meesho.com/women-men-kids-footwear/clp/15mz"
r=requests.get(url)
print(r)
```

```python
soup=BeautifulSoup(r.text,"lxml")
print(soup)
```

```python
Product_Name=[]
names=soup.find_all("p",class_="sc-eDvSVe gQDOBc NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ejhQZU NewProductCardstyled__StyledDesktopProductTitle-sc-6y2tys-5 ejhQZU")
for i in names:
    name=i.text
    Product_Name.append(name)
    print(Product_Name)
```

```python
Price=[]
prices=soup.find_all("h5",class_="sc-eDvSVe dwCrSh")
for i in prices:
    price=i.text
    Price.append(price)
    print(Price)
```

```python
Rating = []
ratings = soup.find_all("span", class_="sc-eDvSVe laVOtN")
for i in ratings:
    rating = i.text.strip()
    Rating.append(rating)
    print(Rating)
```

```python
Review_Count = []
reviews = soup.find_all("span", class_="sc-eDvSVe XndEO NewProductCardstyled__RatingCount-sc-6y2tys-22 iaGtYc NewProductCardstyled__RatingCount-sc-6y2tys-22 iaGtYc")
for i in reviews:
    review = i.text.strip()
    Review_Count.append(review)
    print(Review_Count)
```

```python
df = pd.DataFrame({"product name": Product_Name, "price": Price, "rating": Rating, "review count": Review_Count})
df.to_csv("C:\\Users\\Aniket\\Downloads\\College\\TE-Sem6\\DSBDA_Final\\7_Webscraping\\mesho1.csv")
```
